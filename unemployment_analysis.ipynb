```python
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
```

```python
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv"

urllib.request.urlretrieve(url, "unemployment.csv")

print("Dataset Downloaded Successfully!")
```

```python
df = pd.read_csv("unemployment.csv")

# Remove extra spaces in column names
df.columns = df.columns.str.strip()

df.head()
```

```python
print("Shape of Dataset:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nDataset Info:\n")
print(df.info())
print("\nStatistical Summary:\n")
print(df.describe())
```

```python
avg_rate = df["Estimated Unemployment Rate (%)"].mean()
print("Average Unemployment Rate:", round(avg_rate, 2))
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Estimated Unemployment Rate (%)"])
plt.title("Unemployment Rate Trend")
plt.xlabel("Index")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
region_avg = df.groupby("Region")["Estimated Unemployment Rate (%)"].mean()

print("Region-wise Average Unemployment Rate:\n")
print(region_avg.sort_values(ascending=False))
```
